# Brick Wall

There is a rectangular brick wall in front of you with n rows of bricks. The ith row has some number of bricks each of the same height (i.e., one unit) but they can be of different widths. The total width of each row is the same.

Draw a vertical line from the top to the bottom and cross the least bricks. If your line goes through the edge of a brick, then the brick is not considered as crossed. You cannot draw a line just along one of the two vertical edges of the wall, in which case the line will obviously cross no bricks.

Given the 2D array wall that contains the information about the wall, return the minimum number of crossed bricks after drawing such a vertical line.

 

**Example 1:**



Input: wall = [[1,2,2,1],[3,1,2],[1,3,2],[2,4],[3,1,2],[1,3,1,1]]
Output: 2

**Example 2:**

Input: wall = [[1],[1],[1]]
Output: 3
 

**Constraints:**

- n == wall.length
- 1 <= n <= 104
- 1 <= wall[i].length <= 104
- 1 <= sum(wall[i].length) <= 2 * 104
- sum(wall[i]) is the same for each row i.
- 1 <= wall[i][j] <= 231 - 1

```python
def leastBricks(wall):
    from collections import defaultdict

    # dictionary to keep track of the frequencies of brick edges
    edges = defaultdict(int)

    for row in wall:
        sum_ = 0
        # note that we don't consider the last brick's edge, hence the "-1"
        for brick in row[:-1]:
            sum_ += brick
            edges[sum_] += 1

    # if there are no common edges, we return the number of rows (which means the line crosses all bricks)
    if not edges:
        return len(wall)

    # return the total rows minus the most frequent edge
    return len(wall) - max(edges.values())

# Test cases
wall1 = [[1,2,2,1],[3,1,2],[1,3,2],[2,4],[3,1,2],[1,3,1,1]]
wall2 = [[1],[1],[1]]

print(leastBricks(wall1))  # Output: 2
print(leastBricks(wall2))  # Output: 3
```
