# N-Queens II

The n-queens puzzle is the problem of placing n queens on an n x n chessboard such that no two queens attack each other.

Given an integer n, return the number of distinct solutions to the n-queens puzzle.

 

**Example 1:**


Input: n = 4
Output: 2
Explanation: There are two distinct solutions to the 4-queens puzzle as shown.

**Example 2:**

Input: n = 1
Output: 1
 

Constraints:

1 <= n <= 9

```python
class Solution:
    def totalNQueens(self, n: int) -> int:
        # Initialize a list to store the positions of queens in each column
        queens = [-1] * n
        # Call the backtracking function
        return self.backtrack(queens, 0, n)

    def backtrack(self, queens, row, n):
        # If all queens have been placed, return 1 as a solution is found
        if row == n:
            return 1

        count = 0
        for col in range(n):
            if self.is_valid(queens, row, col, n):
                queens[row] = col
                count += self.backtrack(queens, row + 1, n)
                queens[row] = -1
        return count

    def is_valid(self, queens, row, col, n):
        # Check if a queen can be placed in a given position
        for r in range(row):
            if queens[r] == col or abs(queens[r] - col) == abs(r - row):
                return False
        return True
```

```python
# Initialize the Solution class
s = Solution()

# Test the totalNQueens function
n = 4
print(s.totalNQueens(n))
# Output: 2
```
