# Largest Number

Given a list of non-negative integers nums, arrange them such that they form the largest number and return it.

Since the result may be very large, so you need to return a string instead of an integer.
 
**Example 1:**

Input: nums = [10,2]
Output: "210"

**Example 2:**

Input: nums = [3,30,34,5,9]
Output: "9534330"

**Constraints:**

- 1 <= nums.length <= 100
- 0 <= nums[i] <= 109

```python
class Solution:
    def largestNumber(self, nums):
        # Convert each number to string
        nums = list(map(str, nums))
        
        # Custom sort: descending by concatenation result
        nums.sort(key=lambda x: x*10, reverse=True)
        
        # Alternative (more explicit comparator using functools.cmp_to_key):
        # nums.sort(key=functools.cmp_to_key(lambda a, b: (a+b > b+a) - (a+b < b+a)), reverse=True)

        # Edge case: if the largest is "0", all are zeros
        if nums[0] == "0":
            return "0"
        
        return "".join(nums)
```
