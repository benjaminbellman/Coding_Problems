# N-ary Tree Postorder Traversal

Given the root of an n-ary tree, return the postorder traversal of its nodes' values.

Nary-Tree input serialization is represented in their level order traversal. Each group of children is separated by the null value (See examples)

 

**Example 1:**



Input: root = [1,null,3,2,4,null,5,6]

Output: [5,6,3,2,4,1]

**Example 2:**



Input: root = [1,null,2,3,4,5,null,null,6,7,null,8,null,9,10,null,null,11,null,12,null,13,null,null,14]

Output: [2,6,14,11,7,3,12,8,4,13,9,10,5,1]
 

**Constraints:**

- The number of nodes in the tree is in the range [0, 104].
- 0 <= Node.val <= 104
- The height of the n-ary tree is less than or equal to 1000.

```python
class Node:
    def __init__(self, val=None, children=None):
        self.val = val
        self.children = children if children is not None else []

def postorder(root):
    if not root:
        return []

    stack, output = [root], []
    while stack:
        node = stack.pop()
        output.append(node.val)
        stack.extend(node.children)  # Extend to add all children to the stack

    return output[::-1]  # Reverse the output list

# Example 1: [1,null,3,2,4,null,5,6]
root1 = Node(1, [Node(3, [Node(5), Node(6)]), Node(2), Node(4)])

# Example 2: [1,null,2,3,4,5,null,null,6,7,null,8,null,9,10,null,null,11,null,12,null,13,null,null,14]
root2 = Node(1, [Node(2), Node(3, [Node(6), Node(7, [Node(11, [Node(14)])])]), 
                 Node(4, [Node(8, [Node(12)])]), Node(5, [Node(9, [Node(13)]), Node(10)])])

# Get postorder traversal results
result1 = postorder(root1)
result2 = postorder(root2)

print(result1, result2)
```
