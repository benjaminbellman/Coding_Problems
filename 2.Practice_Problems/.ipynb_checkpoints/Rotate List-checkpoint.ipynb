# Rotate List

Given the head of a linked list, rotate the list to the right by k places.

 

**Example 1:**





Input: head = [1,2,3,4,5], k = 2
Output: [4,5,1,2,3]

**Example 2:**



Input: head = [0,1,2], k = 4
Output: [2,0,1]
 

**Constraints:**

The number of nodes in the list is in the range [0, 500].
-100 <= Node.val <= 100
0 <= k <= 2 * 109

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def rotateRight(head: ListNode, k: int) -> ListNode:
    if not head:
        return None
    if not head.next:
        return head

    # Find the length of the linked list
    length = 1
    curr = head
    while curr.next:
        curr = curr.next
        length += 1

    # Connect the tail to the head to make it a circular list
    curr.next = head

    # Find the new tail, which is (length - k % length - 1)th node
    new_tail = head
    for i in range(length - k % length - 1):
        new_tail = new_tail.next

    # The new head is the next node of the new tail
    new_head = new_tail.next

    # Break the circularity
    new_tail.next = None

    return new_head
```

```python
# Test Example 1
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)
k = 2
result = rotateRight(head, k)
print("Expected Output : [4,5,1,2,3]")
print("Output : [",end="")
while result:
    print(result.val,end=",")
    result = result.next
print("]")

# Test Example 2
head = ListNode(0)
head.next = ListNode(1)
head.next.next = ListNode(2)
k = 4
result = rotateRight(head, k)
print("Expected Output : [2,0,1]")
print("Output : [",end="")
while result:
    print(result.val,end=",")
    result = result.next
print("]")
```
