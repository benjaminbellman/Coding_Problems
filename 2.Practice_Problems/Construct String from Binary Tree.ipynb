# Construct String from Binary Tree

Given the root of a binary tree, construct a string consisting of parenthesis and integers from a binary tree with the preorder traversal way, and return it.

Omit all the empty parenthesis pairs that do not affect the one-to-one mapping relationship between the string and the original binary tree.

 

**Example 1:**



Input: root = [1,2,3,4]

Output: "1(2(4))(3)"

Explanation: Originally, it needs to be "1(2(4)())(3()())", but you need to omit all the unnecessary empty parenthesis pairs. And it will be "1(2(4))(3)"

**Example 2:**



Input: root = [1,2,3,null,4]

Output: "1(2()(4))(3)"

Explanation: Almost the same as the first example, except we cannot omit the first parenthesis pair to break the one-to-one mapping relationship between the input and the output.
 

**Constraints:**

- The number of nodes in the tree is in the range [1, 104].
- -1000 <= Node.val <= 1000

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def treeToStr(root):
    if not root:
        return ""

    # Convert the root value to string
    result = str(root.val)

    # If there's a left child, process it and append the result
    if root.left:
        result += "(" + treeToStr(root.left) + ")"
    # If there's no left child but a right child, we need to add an empty parenthesis pair for left
    elif root.right:
        result += "()"

    # If there's a right child, process it and append the result
    if root.right:
        result += "(" + treeToStr(root.right) + ")"

    return result

# Test the function with the provided examples
# Example 1: root = [1,2,3,4]
root1 = TreeNode(1, TreeNode(2, TreeNode(4)), TreeNode(3))
output1 = treeToStr(root1)

# Example 2: root = [1,2,3,null,4]
root2 = TreeNode(1, TreeNode(2, None, TreeNode(4)), TreeNode(3))
output2 = treeToStr(root2)

output1, output2
```
