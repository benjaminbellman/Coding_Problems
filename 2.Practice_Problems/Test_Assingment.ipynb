## Peregrine Problem

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv("C:\\Users\\benja\\OneDrive\\Desktop\\Police Data.csv")

# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
# Basic statistics for numerical columns
numerical_stats = data.describe()

# Transaction amount distribution
plt.figure(figsize=(12, 6))
sns.histplot(data['Amount'], bins=50, kde=True)
plt.title('Distribution of Transaction Amounts')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.show()

# Count of transactions per sender
sender_transaction_counts = data['Sender Name_'].value_counts()

# Count of transactions per receiver
receiver_transaction_counts = data['Payee Name_'].value_counts()

# Visualize the top 10 senders and receivers by transaction count
plt.figure(figsize=(15, 7))
plt.subplot(1, 2, 1)
sender_transaction_counts.head(10).plot(kind='bar')
plt.title('Top 10 Senders by Transaction Count')
plt.xlabel('Sender Name')
plt.ylabel('Number of Transactions')

plt.subplot(1, 2, 2)
receiver_transaction_counts.head(10).plot(kind='bar', color='orange')
plt.title('Top 10 Receivers by Transaction Count')
plt.xlabel('Payee Name')
plt.ylabel('Number of Transactions')

plt.tight_layout()
plt.show()
```

```python
from datetime import datetime

# Convert Send DateTime and Pay DateTime to datetime objects
data['Send DateTime'] = pd.to_datetime(data['Send DateTime'])
data['Pay DateTime'] = pd.to_datetime(data['Pay DateTime'])

# Calculate the time difference between sending and receiving in hours
data['Time Difference (hours)'] = (data['Pay DateTime'] - data['Send DateTime']).dt.total_seconds() / 3600

# Basic statistics for time difference
time_diff_stats = data['Time Difference (hours)'].describe()

# Distribution of time differences
plt.figure(figsize=(12, 6))
sns.histplot(data['Time Difference (hours)'], bins=50, kde=True)
plt.title('Distribution of Time Differences Between Sending and Receiving (in hours)')
plt.xlabel('Time Difference (hours)')
plt.ylabel('Frequency')
plt.show()
```

```python
time_diff_stats
```

```python
# Define the threshold for short time transactions (2 hours)
short_time_threshold = 2

# Identify transactions that are completed within this threshold
short_time_transactions = data[data['Time Difference (hours)'] <= short_time_threshold]

# Number of transactions identified as short time
num_short_time_transactions = short_time_transactions.shape[0]

# Displaying some of these transactions
short_time_transactions_sample = short_time_transactions.head()
```

```python
# Calculate the threshold for suspiciously large transactions
mean_amount = numerical_stats['Amount']['mean']
std_amount = numerical_stats['Amount']['std']
suspicious_amount_threshold = mean_amount + 2 * std_amount

# Identify transactions that are larger than this threshold
suspicious_large_transactions = data[data['Amount'] > suspicious_amount_threshold]

# Number of transactions identified as suspiciously large
num_suspicious_large_transactions = suspicious_large_transactions.shape[0]

# Displaying some of these transactions
suspicious_large_transactions_sample = suspicious_large_transactions.head()
```

```python
# Calculate the threshold for suspiciously high frequency for senders and receivers
mean_sender_freq = sender_transaction_counts.mean()
std_sender_freq = sender_transaction_counts.std()
suspicious_sender_freq_threshold = mean_sender_freq + 2 * std_sender_freq

mean_receiver_freq = receiver_transaction_counts.mean()
std_receiver_freq = receiver_transaction_counts.std()
suspicious_receiver_freq_threshold = mean_receiver_freq + 2 * std_receiver_freq

# Identify senders and receivers that exceed these thresholds
suspicious_senders = sender_transaction_counts[sender_transaction_counts > suspicious_sender_freq_threshold]
suspicious_receivers = receiver_transaction_counts[receiver_transaction_counts > suspicious_receiver_freq_threshold]

# Number of suspicious senders and receivers
num_suspicious_senders = suspicious_senders.shape[0]
num_suspicious_receivers = suspicious_receivers.shape[0]

suspicious_sender_freq_threshold, num_suspicious_senders, suspicious_senders.head(), suspicious_receiver_freq_threshold, num_suspicious_receivers, suspicious_receivers.head()
```
