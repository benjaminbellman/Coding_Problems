# Shortest Palindrome

You are given a string s. You can convert s to a palindrome by adding characters in front of it.

Return the shortest palindrome you can find by performing this transformation.

**Example 1:**

Input: s = "aacecaaa"
Output: "aaacecaaa"

**Example 2:**

Input: s = "abcd"
Output: "dcbabcd"
 
**Constraints:**

- 0 <= s.length <= 5 * 104
- s consists of lowercase English letters only.

```python
class Solution:
    def shortestPalindrome(self, s: str) -> str:
        if not s:
            return s
        
        rev = s[::-1]
        t = s + "#" + rev
        
        # Build LPS array for KMP
        lps = [0] * len(t)
        j = 0  # length of current longest prefix-suffix
        
        for i in range(1, len(t)):
            while j > 0 and t[i] != t[j]:
                j = lps[j - 1]
            if t[i] == t[j]:
                j += 1
                lps[i] = j
        
        pal_len = lps[-1]                 # longest palindromic prefix length
        to_add = s[pal_len:][::-1]        # reverse the leftover suffix
        return to_add + s
```
